##### Run model :

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_squared_error

features = ['Room', 'Area_log', 'Parking', 'Warehouse', 'Address']
target = 'Price_log'

X = df[features].copy()
y = df[target].copy()
```

##### Cross Validation :

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

r2_scores = []
mse_scores = []
```

##### Manually perform Cross Validation to implement Target Encoding correctly :

```python
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index].copy(), X.iloc[test_index].copy()
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Calculating the average price for each address from the training data :
    train_df = X_train.copy()
    train_df['Price_log'] = y_train
    address_mean_price = train_df.groupby('Address')['Price_log'].mean()

    # Applying Target Encoding to Addresses :
    X_train['Address_encoded'] = X_train['Address'].map(address_mean_price)
    X_test['Address_encoded'] = X_test['Address'].map(address_mean_price)

    # Initializing NaNs with the overall average :
    global_mean = y_train.mean()
    X_train['Address_encoded'] = X_train['Address_encoded'].fillna(global_mean)
    X_test['Address_encoded'] = X_test['Address_encoded'].fillna(global_mean)

    # Delete the Address text column : 
    X_train = X_train.drop('Address', axis=1)
    X_test = X_test.drop('Address', axis=1)

    # Model training :
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Prediction and evaluation : 
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    mse = (mean_squared_error(y_test, y_pred))

    r2_scores.append(r2)
    mse_scores.append(mse)
```

##### final results :

```python
print("RÂ² scores:", r2_scores)
print("Average RÂ²:", np.mean(r2_scores))


print("\nMSE scores:", mse_scores)
print("Average MSE:", np.mean(mse_scores))
```
